```python
def get_stats(tokens):
    counts = {}
    for pair in zip(tokens, tokens[1:]):
        counts[pair] = counts.get(pair, 0) + 1
    return counts

stats = get_stats(tokens)
```

```python
print(sorted(((v, k) for k,v in stats.items()), reverse=True))
```

```python
top_pair = max(stats, key=stats.get)
print(top_pair)
```

```python
def merge(ids, pair, idx):
    new_ids = []
    i = 0
    n = len(ids)
    while i<n:
        if i<n-1 and ids[i] == pair[0] and ids[i+1]==pair[1]:
            new_ids.append(idx)
            i += 2
        else:
            new_ids.append(ids[i])
            i+=1
    return new_ids
```

```python
vocab_size = 276
num_merges = vocab_size - 256
ids = list(tokens)


merges = {}
for i in range(num_merges):
    stats = get_stats(ids)
    top_pair = max(stats, key = stats.get)
    idx = 256+i
    ids = merge(ids, top_pair, idx)
    merges[top_pair] = idx
    print(f"merged {top_pair} into {idx}")
```

```python
vocab = {idx: bytes([idx]) for idx in range(256)}
for pair, idx in merges.items():
    vocab[idx] = (vocab[pair[0]] + vocab[pair[1]])
```

```python
def decode(ids):
    items = []
    for idx in ids:
        items.append(vocab[idx])

    return b"".join(items).decode("utf-8", errors="replace")
```

```python
def encode(text: str):
    tokens = list(text.encode("utf-8"))
    if len(text)<=2: return tokens
    while True:
        stats = get_stats(tokens)  # dict of pair freq.
        # check each pair if anything maches with *merges*, if match then replace
        pair = min(stats, key=lambda x: merges.get(x, float("inf")))
        # if there is no match then break
        if pair not in merges:
            break
        
        # return the tokens
        tokens = merge(tokens, pair, merges[pair])
        
    return tokens
        
        
```
